## Cycling:

1) 'for to do'

```python
for i in range(10):
    print(i)
```

2) 'while'

```python
a = 10
while a != 0:
    print(a)
    a = a-1
```

# - Arrays, matrices

## Arrays:

```python
Array = [0,1,2,3,4,]
print(Array)
```

A simple example of program with arrays:

```python
#-----Array input
#Array initialization
Array1 = [0,0,0,0,0]

#Array input cycle
for i in range(5):
    element = input()
    Array1[i] = element

#Gain array output
print(Array1)
```

## Matrices:

```python
Matrice = [ [1,2,3],[4,5,6],[7,8,9] ]
print(Matrice)
```

A simple example of program with matrices:

```python
#-----Matrice input
#Matrice initialization
Matrice = [[0,0,0],[0,0,0],[0,0,0]]

#Matrice input cycle
for i in range(3):
    for j in range(3):
        element_m = input()
        Matrice[i][j] = element_m
        
#Gain matrice output
print(Matrice)
```
